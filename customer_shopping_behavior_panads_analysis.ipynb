```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
# it fills missing Review Rating values with the median rating of their respective category
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ' ,'_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
# this code convert our data into snake casing (lower case)
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young_Adult' , 'Adult' , 'Middle_aged' , 'Senior']
df['age_group'] = pd.qcut(df['age'] , q=4 , labels = labels)
# creates an age_group column by dividing the age data into (qcut) four equal-sized groups
```

```python
df[['age' , 'age_group']].head(5)
```

```python
# create a column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' :14,
    'Weekly' : 7 ,
    'Monthly' : 30,
    'Quarterly' : 90 ,
    'Bi-Weekly': 14,
    'Annually':365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days' , 'frequency_of_purchases']].head(5)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
# value comes true so both colms are equal so we can drop any one colm 
```
